#Pengaruh Jenis Pekerjaan terhadap Pendapatan Berdasarkan Lama Bekerja Menggunakan MANOVA

Dalam bidang ketenagakerjaan, analisis terhadap faktor-faktor yang memengaruhi tingkat pendapatan individu menjadi isu sentral, khususnya dalam penyusunan kebijakan ekonomi dan pengambilan keputusan bisnis. Salah satu variabel yang patut mendapat perhatian adalah jenis pekerjaan yang dijalani seseorang, yang diyakini memiliki korelasi erat dengan besaran pendapatan yang diperoleh, terutama ketika dikaji bersama dengan durasi pengalaman kerja. Realitas di lapangan menunjukkan bahwa perbedaan jenis pekerjaan kerap kali menghasilkan variasi pendapatan, meskipun pengalaman kerja yang dimiliki relatif setara. Pemahaman mendalam mengenai pola ini sangat dibutuhkan oleh perusahaan maupun institusi keuangan dalam rangka mengidentifikasi segmen pekerjaan dengan potensi ekonomi yang tinggi, sehingga strategi pemasaran dan kebijakan kredit dapat disusun secara lebih efektif dan tepat sasaran.

Penelitian ini mengadopsi metode Multivariate Analysis of Variance (MANOVA) untuk menganalisis pengaruh jenis pekerjaan terhadap dua variabel dependen secara bersamaan, yaitu pendapatan bulanan dan lama bekerja. Pendekatan ini memungkinkan peneliti untuk tidak hanya menelaah perbedaan pendapatan antar kelompok pekerjaan, tetapi juga menginvestigasi peran pengalaman kerja dalam memengaruhi perbedaan tersebut. Dengan memanfaatkan data dari konsumen yang mengajukan kredit atau melakukan pembelian kendaraan secara tunai, studi ini diharapkan dapat memberikan gambaran empiris mengenai segmentasi pekerjaan dengan daya beli yang relatif tinggi. Hasil penelitian ini dapat menjadi acuan bagi perusahaan, khususnya di sektor otomotif atau keuangan, dalam merumuskan strategi pemasaran berbasis klasifikasi pekerjaan dan pengalaman kerja yang terbukti berpengaruh terhadap tingkat pendapatan.

Berdasarkan pertimbangan metodologis, penelitian ini mengimplementasikan Multivariate Analysis of Variance (MANOVA) guna mengkaji pengaruh jenis pekerjaan terhadap pendapatan dengan mengontrol variabel lama bekerja. Pemilihan MANOVA didasarkan pada kemampuannya menganalisis dua variabel dependen secara terintegrasi, yaitu pendapatan bulanan (monthly income) dan durasi pengalaman kerja (years of employment), terhadap satu variabel independen (jenis pekerjaan). Metode ini tidak sekadar mengidentifikasi perbedaan pendapatan antarkategori pekerjaan, tetapi juga memungkinkan peneliti mengeksplorasi peran pengalaman kerja dalam memediasi hubungan antar-variabel secara multivariat sebuah analisis yang tidak tercakup dalam uji ANOVA konvensional.

Dalam konteks pemasaran dan evaluasi kapasitas ekonomi konsumen, pendekatan ini menjadi krusial untuk menyusun pemetaan komprehensif mengenai kelompok pekerjaan dengan kecenderungan daya beli superior, yang selanjutnya dapat dijadikan dasar objektif dalam segmentasi pasar. Keunggulan MANOVA terletak pada kemampuannya meminimalkan bias analitik melalui pertimbangan korelasi antar-variabel dependen dan interaksi antar-faktor, berbeda dengan metode univariat yang berpotensi mengabaikan kompleksitas hubungan antar-variabel.

Implikasi praktis dari pendekatan ini mencakup dua aspek: pertama, memberikan kontribusi teoretis dalam memahami dinamika ketenagakerjaan; kedua, menjadi acuan operasional bagi perusahaan untuk merancang strategi pemasaran berbasis klasifikasi pekerjaan dan pengalaman kerja. Hasil analisis dapat dioptimalkan untuk menyusun skema kredit atau kampanye penjualan kendaraan yang lebih presisi, efisien, dan sesuai dengan profil ekonomi target konsumen.

Penelitian ini berangkat dari asumsi bahwa jenis pekerjaan dan durasi bekerja merupakan faktor yang berkorelasi secara intrinsik dan secara kolektif memengaruhi tingkat pendapatan individu. Asumsi ini menggarisbawahi bahwa disparitas pendapatan antarkelompok pekerjaan tidak semata-mata dipengaruhi oleh karakteristik profesi, melainkan juga oleh akumulasi pengalaman kerja dalam bidang tersebut. Melalui penerapan MANOVA (Multivariate Analysis of Variance), studi ini mengasumsikan adanya interdependensi antara dua variabel dependen pendapatan bulanan dan lama bekerja - sehingga analisis multivariat dipandang lebih relevan ketimbang pendekatan univariat yang mengabaikan dinamika hubungan antar-variabel.

Lebih lanjut, penelitian ini mengasumsikan bahwa data historis konsumen (seperti riwayat pengajuan kredit atau pembelian kendaraan) mampu merepresentasikan perilaku ekonomi berdasarkan profil pekerjaan dan pengalaman kerja, yang pada gilirannya mencerminkan estimasi daya beli. Dengan demikian, MANOVA dipilih sebagai metode analisis karena kemampuannya mengakomodasi kompleksitas variasi pendapatan secara multidimensi berdasarkan klasifikasi pekerjaan, sekaligus meminimalkan bias yang mungkin timbul dari pendekatan analitik terpisah terhadap masing-masing variabel dependen. Melibatkan serangkaian prosedur untuk menguji adanya pengaruh signifikan jenis pekerjaan terhadap pendapatan bulanan individu dengan mempertimbangkan durasi bekerja, menggunakan metode Multivariate Analysis of Variance (MANOVA). Tujuan utama analisis ini adalah untuk mengeksplorasi pola hubungan multivariat antara kategori pekerjaan dengan dua variabel dependen, yaitu pendapatan bulanan (Monthly Income) dan lama bekerja (Years of Employment).

```python
from sklearn.preprocessing import StandardScaler

# Kolom numerik yang akan dianalisis
numeric_cols = ['Monthly Income', 'Years of Employment']
data_transformed = df[numeric_cols].copy()

# Buang nilai NaN
data_transformed = data_transformed.dropna()

# Tangani nilai nol atau negatif (tidak valid untuk log)
epsilon = 1e-6
for col in numeric_cols:
    min_val = data_transformed[col].min()
    if min_val <= 0:
        data_transformed[col] = data_transformed[col] + abs(min_val) + epsilon

# Transformasi log10
for col in numeric_cols:
    data_transformed[col] = np.log10(data_transformed[col])

# Standardisasi (z-score)
scaler = StandardScaler()
data_transformed[numeric_cols] = scaler.fit_transform(data_transformed[numeric_cols])

# Lihat hasil transformasi
print(data_transformed.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import QuantileTransformer
from scipy.stats import zscore
import pingouin as pg

# Step 1: Siapkan DataFrame awal
df_transformed_full = df.copy()

# Pastikan Monthly Income numerik
df_transformed_full['Monthly Income'] = (
    df_transformed_full['Monthly Income']
    .astype(str)
    .str.replace('.', '', regex=False)
    .str.replace(',', '.', regex=False)
)
df_transformed_full['Monthly Income'] = pd.to_numeric(df_transformed_full['Monthly Income'], errors='coerce')

# Buat Credit Score dummy jika belum ada
if 'Credit Score' not in df_transformed_full.columns:
    np.random.seed(42)
    df_transformed_full['Credit Score'] = (
        df_transformed_full['Monthly Income'] * 0.01 +
        np.random.normal(650, 30, len(df_transformed_full))
    )

# Hapus NaN
df_transformed_full = df_transformed_full.dropna(subset=['Monthly Income', 'Years of Employment'])

# Step 2: Transformasi Quantile + z-score
qt = QuantileTransformer(output_distribution='normal', random_state=42)
transformed_array = qt.fit_transform(df_transformed_full[['Monthly Income', 'Years of Employment']])
df_transformed_full['Monthly Income_transformed'] = zscore(transformed_array[:, 0])
df_transformed_full['Years of Employment_transformed'] = zscore(transformed_array[:, 1])

# Step 3: Buang outlier dengan z-score
df_homogen = df_transformed_full.copy()
df_homogen['z_income'] = zscore(df_homogen['Monthly Income_transformed'])
df_homogen['z_credit'] = zscore(df_homogen['Years of Employment_transformed'])

df_homogen = df_homogen[
    (df_homogen['z_income'].abs() < 3) &
    (df_homogen['z_credit'].abs() < 3)
]
df_homogen = df_homogen.drop(columns=['z_income', 'z_credit'])

# Step 4: Uji Box’s M
result = pg.box_m(
    data=df_homogen,
    dvs=['Monthly Income_transformed', 'Years of Employment_transformed'],
    group='Bidang_Kategori'  # Ganti ke 'Occupation_Kategori' jika kamu belum kategorikan
)

print(result)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.preprocessing import PowerTransformer
import pingouin as pg

# Step 1: Salin data dan pastikan kolom numerik
df_clean = df.copy()
df_clean['Monthly Income'] = (
    df_clean['Monthly Income']
    .astype(str)
    .str.replace('.', '', regex=False)
    .str.replace(',', '.', regex=False)
)
df_clean['Monthly Income'] = pd.to_numeric(df_clean['Monthly Income'], errors='coerce')

# Buat dummy Credit Score jika belum ada
if 'Years of Employment' not in df_clean.columns:
    np.random.seed(42)
    df_clean['Credit Score'] = df_clean['Monthly Income'] * 0.01 + np.random.normal(650, 30, len(df_clean))

# Hapus baris kosong
df_clean = df_clean.dropna(subset=['Monthly Income', 'Years of Employment', 'Occupation_Kategori'])

# Step 2: Gabungkan kategori yang jumlahnya sedikit jadi 'Lainnya'
kategori_counts = df_clean['Occupation_Kategori'].value_counts()
minor_kategori = kategori_counts[kategori_counts < 15].index.tolist()

df_clean['Occupation_Kategori'] = df_clean['Occupation_Kategori'].replace(minor_kategori, 'Lainnya')

# Step 3: Transformasi Yeo-Johnson (lebih fleksibel dari log)
pt = PowerTransformer(method='yeo-johnson')
numeric_cols = ['Monthly Income', 'Years of Employment']
transformed = pt.fit_transform(df_clean[numeric_cols])

df_clean['Monthly Income_transformed'] = zscore(transformed[:, 0])
df_clean['Credit Score_transformed'] = zscore(transformed[:, 1])

# Step 4: Buang outlier Z-score > 2.5
df_clean['z_income'] = zscore(df_clean['Monthly Income_transformed'])
df_clean['z_credit'] = zscore(df_clean['Years of Employment_transformed'])

df_homogen = df_clean[(df_clean['z_income'].abs() < 2.5) & (df_clean['z_credit'].abs() < 2.5)].copy()

# Hapus kolom z bantu
df_homogen = df_homogen.drop(columns=['z_income', 'z_credit'])

# Step 5: Uji Box's M
result = pg.box_m(
    data=df_homogen,
    dvs=['Monthly Income_transformed', 'Years of Employment_transformed'],
    group='Occupation_Kategori'
)

print(result)
```

```python
from statsmodels.multivariate.manova import MANOVA

# Monthly_Income dan Years_of_Employment = variabel dependen
# Occupation = variabel independen
manova = MANOVA.from_formula('Monthly_Income + Years_of_Employment ~ Occupation', data=df_bersih)
print(manova.mv_test())
```

Wilks’ Lambda mengukur proporsi varians dalam variabel dependen yang tidak dapat dijelaskan oleh variabel independen.
